```python
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
# judel_model_names = ['juv_1790_1830', '1830', '1840', '1850', '1860', '1870', '1880']
judel_model_names = ['juv_1790_1830'] + [str(i) for i in range(1830, 2001, 10)]
k = 11

knn_ddicts = get_knn_ddicts('JUDEL', judel_model_names, k)
```

# raw JUDEL results

```python
judel_y_vals = list()
for i in range(len(judel_model_names) - 1):
    juv_knn_ddict_c1 = knn_ddicts[judel_model_names[i]]
    juv_knn_ddict_c2 = knn_ddicts[judel_model_names[i+1]]
    judel_y_vals.append(knn_ccla(juv_knn_ddict_c1, juv_knn_ddict_c2))
judel_y_vals = np.array(judel_y_vals)
x_vals = [1823]   # round( (((1790 + 1830) / 2) + 1835) / 2 )
x_vals += [str(i) for i in range(1840, 2001, 10)]  # start at (1835 + 1845) / 2 = 1840
```

```python
fig, ax = plt.subplots()
ax.plot(x_vals, judel_y_vals)
ax.set_title('JUDEL CCLA time series')
ax.set_xticks(ax.get_xticks()[::2])
fig.show()
```

# random average time series

```python
models = [load_model(name) for name in judel_model_names]

# get 100 random words
rand_words = list()
while len(rand_words) < 100:
    include = True
    word = random.choice(models[0].index_to_key)
    for model in models:
        if not model.has_index_for(word):
            include = False
            break
    if include:
        rand_words.append(word)
del models
```

```python
all_w_y_vals = list()
for word in rand_words:
    w_knn_ddicts = get_knn_ddicts(word, judel_model_names, k)
    w_y_vals = list()
    for i in range(len(judel_model_names) - 1):
        w_ddict_c1 = w_knn_ddicts[judel_model_names[i]]
        w_ddict_c2 = w_knn_ddicts[judel_model_names[i + 1]]
        w_y_vals.append(knn_ccla(w_ddict_c1, w_ddict_c2))
    all_w_y_vals.append(w_y_vals)
```

```python
all_w_y_vals = np.array(all_w_y_vals)
avg_y_vals = np.mean(all_w_y_vals, axis=0)
```

# normalized results for known semantic-shift terms

```python
def tmp_ccla_y_vals(word):
    term_knn_ddicts = get_knn_ddicts(word, judel_model_names, k)
    term_y_vals = list()
    for i in range(len(judel_model_names) - 1):
        term_ddict_c1 = term_knn_ddicts[judel_model_names[i]]
        term_ddict_c2 = term_knn_ddicts[judel_model_names[i + 1]]
        term_y_vals.append(knn_ccla(term_ddict_c1, term_ddict_c2))
    term_y_vals = np.array(term_y_vals)
    return term_y_vals

semshift_terms = ['guy', 'gay', 'broadcast', 'monitor']
semshift_y_vals = dict()
for term in semshift_terms:
    semshift_y_vals[term] = tmp_ccla_y_vals(term)
```

```python
for term, y_vals in semshift_y_vals.items():
    t_diff_series = y_vals - avg_y_vals
    t_diff_series /= np.linalg.norm(t_diff_series)
    t_ratio_series = y_vals / avg_y_vals
    t_ratio_series /= np.linalg.norm(t_ratio_series)

    fig, ax = plt.subplots()
    ax.plot(x_vals, t_diff_series)
    ax.plot(x_vals, t_ratio_series)
    ax.legend(['subtract', 'divide'])
    ax.set_title(f'normalized CCLA time series: {term} (OP) avg')
    ax.set_xticks(ax.get_xticks()[::2])
    fig.show()
```

Expected (qua [Hamilton et al., 2016](https://aclanthology.org/P16-1141.pdf), table 2):
* guy: circa 1850
* gay: circa 1920
* broadcast: circa 1920
* monitor: circa 1930

My interpretation of these results:
* "broadcast" is spot on
* "guy" and "gay" CCLA scores do noticeably drop around where they are expected to, but have additional drops in unexpected places
    * In the "guy" case, it looks like there was broad meaning change around the expected time, but that this meaning gradually stabilized
    * In the "gay" case, it looks like there is a distinct dip at the expected time, but that there are even more pronounced dips in 1870 and 1970-1980. I assume the latter drop corresponds to coverage of the gay liberation movement (which occurred in almost exactly the period we see the dip)
* "monitor" looks like noise

# which words track JUDEL best?

```python
import math
from sklearn.metrics import mean_squared_error

min_sq_err = math.inf
mse_word = ''
mse_y_vals = None
for word in rand_words:
    w_knn_ddicts = get_knn_ddicts(word, judel_model_names, k)
    w_y_vals = list()
    for i in range(len(judel_model_names) - 1):
        w_ddict_c1 = w_knn_ddicts[judel_model_names[i]]
        w_ddict_c2 = w_knn_ddicts[judel_model_names[i + 1]]
        w_y_vals.append(knn_ccla(w_ddict_c1, w_ddict_c2))
    w_y_vals = np.array(w_y_vals)

    sq_err = mean_squared_error(judel_y_vals, w_y_vals)
    if sq_err < min_sq_err:
        min_sq_err = sq_err
        mse_word = word
        mse_y_vals = w_y_vals
```

```python
diff_mse = mse_y_vals - avg_y_vals
diff_mse /= np.linalg.norm(diff_mse)
ratio_mse = mse_y_vals / avg_y_vals
ratio_mse /= np.linalg.norm(ratio_mse)

fig, ax = plt.subplots()
ax.plot(x_vals, diff_series)
ax.plot(x_vals, ratio_series)
ax.plot(x_vals, diff_mse)
ax.plot(x_vals, ratio_mse)
ax.legend(['JUDEL - avg', 'JUDEL / avg', f'{mse_word} - avg', f'{mse_word} / avg'])
ax.set_title(f'normalized CCLA time series: JUDEL vs {mse_word}')
ax.set_xticks(ax.get_xticks()[::2])
fig.show()
```

# Concern (OLD, from 2021.09.09 -- revisit later):
Change in JUDEL CCLA scores over time is correlated with embedding quality. Higher-quality embeddings make for more consistent, reliable similarity metrics (e.g. cosine similarity, the metric used in this CCLA measurement), which lead to higher CCLA scores. This obfuscates whatever CCLA could otherwise tell us.

Embedding quality improves over time (at least in the period examined so far), because:
* Later years have more data.
* Later years have fewer typos (anecdotally, qua me).
 * We could, of course, develop a more rigorous method for quantifying the likely proportion of typos per year, but I'm assuming that this will not be a priority until we've developed better ways to correct/eliminate typos. Thoughts?

A data-limited approach (e.g. taking the number of years/months required to have the same number of articles/words) would help solve the first problem. (I do not have an idea for a solution to the second problem.) However, depending on where we set the threshold here, this would lead to one of the following:
* High threshold: early embeddings would require several decades of data (e.g. 1790-1860 has about the same number of articles as (01.01.)1910-(01.01.)1912, ~1.5M articles).
* Low threshold: all embeddings are lower quality, so while this would bring their CCLA scores more in line with each other without compressing several decades together, overall results would just be less reliable.

We could test out both and see how useful they'd be? The second seems like a better option -- if it ended up being super noisy, we'd at least be able to tell and move on -- but it would also take much more compute time.
