```python
from gensim.models import LdaModel
from gensim.corpora.mmcorpus import MmCorpus
from gensim.models import  ldamulticore
import logging
import gensim
import pandas as pd
import numpy as np
from gensim import corpora
from pprint import pprint
import json
from punctuator import Punctuator
from nltk.tokenize import sent_tokenize
import os
import re
p = Punctuator('/Users/bhavya/.punctuator/Demo-Europarl-EN.pcl')
# logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
min_prob = 0.0
model_file = '../data/juvenile_1830_20.lda'
model = LdaModel.load(model_file)
id2word = model.id2word
corpus_file = '../data/juvenile_1830_corp.txt'
num_topics = 20
corpus = MmCorpus(corpus_file)


# path = '/Users/bhavya/Documents/juvenile/data/juvenile_1830.csv'
# df = pd.read_csv(path)
# df['date'] = pd.to_datetime(df['date'],format="%b %d, %Y @ %H:%M:%S.%f")

# df['year'] = pd.DatetimeIndex(df['date']).year
# yr = df['year'].to_list()
# tle = df['title'].to_list()
# ft = df['fulltext'].to_list()


date_path = '../data/juvenile_1830_date.txt'
title_path = '../data/juvenile_1830_title.txt'
ft_path = '../data/juvenile_1830.txt'

def read(path):
    with open(path) as f:
        data = f.read()
        lst = data.split('\n')
    return lst

dates = read(date_path)
tle = read(title_path)
ft = read(ft_path)

yr = []
for d in dates:
    yr.append(d[:4])



topic_dist_lst = {}
for idx,d in enumerate(corpus):
    lst = model.get_document_topics(d,minimum_probability=min_prob)
    y = yr[idx]
#     print(lst,d)
    for t,prob in lst:
        try:
            topic_dist_lst[y][t].append(prob)
        except:
            try:
                topic_dist_lst[y][t] = [prob]
            except:
                topic_dist_lst[y] = {t:[prob]} 
topic_dist = {}
for y,tp in topic_dist_lst.items():
    for t,p in tp.items():
        try:
            topic_dist[y][t] = np.average(np.array(p))
        except:
            topic_dist[y] = {t:np.average(np.array(p))}
          
            
 

top_topics = model.top_topics(corpus,dictionary=id2word, topn=30)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
# avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
# print('Average topic coherence: %.4f.' % avg_topic_coherence)
pprint(top_topics)    
```

```python
import matplotlib.pyplot as plt
num_topics=20
for i in range(num_topics):
    y = []
    x = []
    for year, tp in sorted(topic_dist.items()):
        x.append(int(year))
        try:
            y.append(tp[i])
        except:
            y.append(0)
    plt.figure(figsize=(10,7))
    plt.plot(x,y,marker='o')
    plt.xlabel('Year')
    plt.ylabel('Average topic probability for articles')
    plt.title( 'Topic '+str(i)+':  '+', '.join([w[0] for w in model.show_topic(i,topn=20)]))
#     plt.legend(bbox_to_anchor=(1, 1.20))
#     plt.show() 
#     plt.tight_layout()
    plt.savefig('../data/'+str(i),bbox_inches='tight')
```

```python
from nltk.tokenize import sent_tokenize
all_sents = {}
bow_sents = {}



for idx,d in enumerate(ft):
#     print(d)
    
    try:
        sents = re.split(r"[.?]",p.punctuate(d))
        all_sents[idx] = sents
    except Exception as e:
        print(e)
        all_sents[idx] = sent_tokenize(d)
        
#     os.mkdir(mallet_sents_path+str(idx))
#     ps = []
    print(idx)
#     for idx2,s in enumerate(sents):

#         ps.append(s.split())
#         with open(mallet_sents_path+str(idx)+'/'+str(idx2)+'.txt','w') as f:
#             f.write(pr_s)
#     BoW_corpus = [id2word.doc2bow(doc, allow_update=False) for doc in ps]
#     bow_sents[idx] = BoW_corpus
```

```python
for idx,sents in all_sents.items():
    ps = []
    print(idx)
    for idx2,s in enumerate(sents):

        ps.append(s.split())
    BoW_corpus = [id2word.doc2bow(doc, allow_update=False) for doc in ps]
    bow_sents[idx] = BoW_corpus
```

```python
lsts = {}
lsts_sents = {}
min_prob1 = 0.01
min_prob2 = 0.01

for idx,d in enumerate(corpus):
    lst = model.get_document_topics(d,minimum_probability=min_prob1)
    lsts[idx] = lst
    lsts_sents[idx] = {}
    print(idx)
    for idxs,s in enumerate(bow_sents[idx]):
        lsts_sents[idx][idxs] = model.get_document_topics(s,minimum_probability=min_prob2)
    
```

```python
for tno in range(num_topics):
    topic_docs = []
    dt_start = 1790
    dt_end = 1830
    num_sents = 10
    tpath = '../data/juvenile_1830_20topics/topic'+str(tno)
    if not os.path.exists(tpath):
        os.mkdir(tpath)
    with open(tpath+'/top_words.txt','w') as f:
        f.write('Word'+'\t'+'Probability'+'\n')
        for t,p in model.show_topic(tno,topn=20):
            f.write(t+'\t'+str(p)+'\n')
    
    for idx,d in enumerate(corpus):
        lst = lsts[idx]
    #     print(lst,d)
        for t,prob in lst:
            if t==tno and int(yr[idx])>=dt_start and int(yr[idx])<=dt_end:
                topic_docs.append((tle[idx],all_sents[idx],dates[idx],prob,idx))
    # print(len(topic_docs))
    topic_docs.sort(key=lambda tup: tup[3],reverse=True)
#     print("Total docs with "+str(min_prob1)+" min topic coverage: ",len(topic_docs))
    print(tno)
    
    for idx,info in enumerate(topic_docs):
        topic_sents = []
        orig_idx = info[4]
        for idxs,_ in enumerate(bow_sents[orig_idx]):
            lst_s = lsts_sents[orig_idx][idxs] 
    
            for t,prob in lst_s:
                if t==tno:
                    topic_sents.append((all_sents[orig_idx][idxs],prob))
        topic_sents.sort(key=lambda tup: tup[1],reverse=True)
        with open(tpath+'/Article'+str(idx)+'.txt','w') as f:
            f.write('Topic probability for article'+': '+str(info[3]))
            f.write('\n---------------------------------------\n')
            f.write('Title: '+info[0])
            f.write('\n---------------------------------------\n')
            f.write('Date: '+info[2])
            f.write('\n---------------------------------------\n')
            f.write('Top '+str(num_sents)+' most relevant sentences for topic:\n')
            for idxs2,si in enumerate(topic_sents[:num_sents]):
                f.write('\t Sentence '+str(idxs2+1)+': '+si[0])
                f.write('\n')
                f.write('\t Topic Probability for sentence: '+str(si[1]))
                f.write('\n')
            f.write('\n---------------------------------------\n')
            f.write('Article: \n\t'+'.\n\t'.join(info[1]))
```
